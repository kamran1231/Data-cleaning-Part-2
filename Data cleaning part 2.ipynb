```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/file/d/1MuZdDAAcIizXVAejicMhzFGzrBO4b1Vu/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)

df.shape
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df.head(6)
```

```python
df.isnull().sum()
```

```python
missing_value_per = df.isnull().sum()/df.shape[0] * 100

missing_value_per
```

```python
missing_value_clm_gre_20 = missing_value_per[missing_value_per >20].keys()

missing_value_clm_gre_20
```

```python
df2_drop_clm = df.drop(columns= missing_value_clm_gre_20)

df2_drop_clm.shape
```

```python

df3_num = df.select_dtypes(include=['int64','float64'])

df3_num.head(6)
```

```python
plt.figure(figsize=(25,25))

sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum().sum()
```

```python
#if you want to access  missing value variable name from numerical variable table


missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum()> 0]

missing_num_var
```

```python
plt.figure(figsize=(10,10))

sns.set()

for i , var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins= 20, kde_kws={'linewidth':5,'color':'#DC143C'})
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean())

df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))

for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins= 20 , kde_kws= {'linewidth': 5,'color':'red'},label='Original')
    sns.distplot(df4_num_mean[var], bins= 20, kde_kws= {'linewidth': 5, 'color': 'green'},label='Mean data')
    plt.legend()
```

```python
df5_num_median = df3_num.fillna(df3_num.median())

df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))

sns.set()

for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins= 20,hist= False, kde_kws={'linewidth': 5, 'color':'red'},label='Original')
    sns.distplot(df4_num_mean[var], bins= 20,hist= False, kde_kws= {'linewidth': 4, 'color': 'green'},label='Mean data')
    sns.distplot(df5_num_median[var],bins= 20,hist= False, kde_kws= {'linewidth': 3, 'color': 'blue'}, label= 'Median')
    plt.legend()
```
